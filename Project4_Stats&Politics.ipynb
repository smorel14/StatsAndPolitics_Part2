To determine if a column is correlated with the Republican vote share we use the following 3 segmentations. <br>
<ul>
  <li>Equal to 0 : No relationship</li>
  <li>Bellow 0.3: Weak relationship</li>   
  <li>Between 0.3 and 0.5: Moderate relationship</li>
  <li>Above 0.5: Strong relationship</li>
</ul>

These 3 segments are in absolute value terms. These segmentations seem to be common guide lines. Both 'Statistics for Dummies' and an article published by IARS ('Correlation Coefficients: Appropriate Use and Interpretation') use this segmentation to determine the strength of the relationship. For this reason we will use the same definition.<br>
</div>

```python
# Correlation for 'Republicans 08, 12 & 16 pct'
corr_frac_dem_08 = df_votes_data.loc[:,:'Injury.deaths'].corrwith(df_votes_data['Republicans 08 pct'])
corr_frac_dem_12 = df_votes_data.loc[:,:'Injury.deaths'].corrwith(df_votes_data['Republicans 12 pct'])
corr_frac_dem_16 = df_votes_data.loc[:,:'Injury.deaths'].corrwith(df_votes_data['Republicans 16 pct'])

# Creating a DataFrame with all 3 series together
df_corr = pd.DataFrame()
df_corr['corr_frac_dem_08'] = pd.Series(corr_frac_dem_08)
df_corr['corr_frac_dem_12'] = pd.Series(corr_frac_dem_12, index=df_corr.index)
df_corr['corr_frac_dem_16'] = pd.Series(corr_frac_dem_16, index=df_corr.index)
df_corr['corr_frac_dem_16_abs'] = pd.Series(corr_frac_dem_16, index=df_corr.index).abs()

# Sorting the DataFrame by the magnitude of the correlation
df_corr = df_corr.sort_values(by=['corr_frac_dem_16_abs'], ascending=False)
display(df_corr.loc[:, :'corr_frac_dem_16'].style.background_gradient())
```

    Additionally we can observe that over the last 3 elections the correlation coefficient has become stronger (both negatively and positively (with the exception of unemployment). How are the racial demographics correlated with Republican vote share? Is there any race which is consistently correlated with the Republican vote share? Are these correlations statistically significant?


<div style = "background-color: #ddffff" >
From the table above we saw that the 'White (Not Latino) Population' and the 'White Asian' population have a strong positive correlation with Republican vote share in 2016. Whiles the African American population and the Asian American population have a moderate (close to strong) negative correlation with Republican vote share in 2016.<br><br>
    To test is these correlations are statistically significant we use the pearson correlation coefficient (Perarson's r) as a measure of the linear correlation between the 2 variables.<br>
    According to the table, as all our variables have p = 0.0 < 0.05, we reject our null hypothesis that the population correlation is equal to 0 for any of our variables.
    
</div>

```python
#To make life easier for the next few questions we have created a function to calculate the correlation and the pearson-r p value.
from scipy import stats

# Function of create a correlation DF.
def corr_info(df, y_value):
    corr_dem = pd.DataFrame()
    names = []
    corr = []
    prob = []

    for name in df.columns:
        typ = df[name].dtypes
        if (typ == float) or (typ == int):
            names.append(name)
            
            # correlation between X and Y
            X = df[name]
            Y = df[y_value]
            corr.append(np.corrcoef(X,Y)[0,1])
            
            # p_value that the correlation between X and Y is not significant
            r, p_value = stats.pearsonr(X, Y)
            prob.append(p_value)
    
    #Saving the name, corelation, and p_value to the DF corr_dem
    corr_dem['name'] = names
    corr_dem['corr'] = np.around(corr,2)
    corr_dem['p-value'] = np.around(prob,4)
    
    #Sorting by the correlations by absolute value
    corr_dem['corr_abs'] = np.absolute(corr_dem['corr'])
    corr_dem = corr_dem.sort_values(by ='corr_abs', ascending=False)
    
    return corr_dem.loc[:,('name','corr','p-value')]

```

```python
# Create a DF with races and republican pct.
df_votes_race = df_votes_data.iloc[:,12:18]
df_votes_race['White_Asian'] = df_votes_data['White_Asian']
df_votes_race['Republicans 08 pct'] = df_votes_data['Republicans 08 pct']
df_votes_race['Republicans 12 pct'] = df_votes_data['Republicans 12 pct']
df_votes_race['Republicans 16 pct'] = df_votes_data['Republicans 16 pct']
```

```python
#Joining the 3 following DF together.
corr_pct_08 = corr_info(df_votes_race, 'Republicans 08 pct')
corr_pct_12 = corr_info(df_votes_race, 'Republicans 12 pct')
corr_pct_16 = corr_info(df_votes_race, 'Republicans 16 pct')


corr_pct = corr_pct_08.join(corr_pct_12.loc[:,('corr','p-value')], lsuffix="_08", rsuffix="_12")\
            .join(corr_pct_16.loc[:,('corr','p-value')])

display(corr_pct.iloc[3:])
```

Look back at the non-aggregated county-level data. Conduct one or more statistical tests to see whether the mean Republican party vote share across all of the counties are different for the 3 elections (Which statistical test is appropriate here? Assume that the people who voted in each county are a representative sample of its population.) Describe why you’ve chosen your tests and interpret your results.

<div style = "background-color: #ddffff" >
Assuming that our data is normally distributed, we conduct an ANOVA test as it is used to test the mean between groups of numerical data. Our null hypothesis is to test whether the mean of all 3 elections are equal. From our analysis bellow we obtain a p-value from our F-statistic well bellow 0.05 and thus reject the null hypothesis. From this we can conclude that the mean Republican party vote share across all the counties are not the same for all 3 elections.<br>
    
To check which elections are different from the rest we conducted Tukey's range test. From the table at the bottom it's clear that we can reject all 3 combinations of the 3 elections at a confidence interval of 0.05. The p-value for each combination is 0.001 < 0.05. Thus we can conclude that the mean Republican party vote share across all the counties are different for all 3 elections. Finally, our analysis also tells us that there is a 3.03% increase in mean Republican party vote share across all the counties between 2008 and 2012. It also shows that there was a 5.93% increase in mean Republican party vote share across all the counties between 2012 and 2016.<br>    

</div>

```python
# Anova
df_pct_data = df_votes_data.loc[:,('Republicans 08 pct','Republicans 12 pct','Republicans 16 pct')]
#display(df_pct_data)
df_pct_data = pd.melt(df_pct_data)
display(df_pct_data.head())
```
